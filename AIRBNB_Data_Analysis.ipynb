```python
#keeping only the columns that we need
#-id, city, neighborhood, price, review_scores_rating, property_type, number_of_reviews
listings_raw = listings_raw.select('id', 'neighbourhood', 'price', 'review_scores_rating', 'property_type', 'number_of_reviews', 'city')
```

a) Basic cleaning of listings

```python
#listings - first cleaning

#the city attribute is crucial to the information, and from the data that we have there is no recover it,
#therefore, we are dropping the records that have null as city
listings_raw = listings_raw.filter(f.col('city').isNotNull())

#the price attribute is crucial to the information, and from the data that we have there is no recover it,
#therefore, we are dropping the records that have null as price
listings_raw = listings_raw.filter(f.col('price').isNotNull())

#the rating attribute is crucial to the information, and from the data that we have there is no recover it,
#therefore, we are dropping the records that have null as rating
listings_raw = listings_raw.filter(f.col('review_scores_rating').isNotNull())

#the id attribute is crucial to the information, and from the data that we have there is no recover it,
#therefore, we are dropping the records that have null as id
listings_raw = listings_raw.filter(f.col('id').isNotNull())

```

b) Basic cleaning of reviews

```python
#reviews - first cleaning

#the id attribute is crucial to the information, and from the data that we have there is no recover it,
#therefore, we are dropping the records that have null as id
reviews_raw = reviews_raw.filter(f.col('listing_id').isNotNull())

#the comment attribute is crucial to the information, and from the data that we have there is no recover it,
#therefore, we are dropping the records that have null as comment
reviews_raw = reviews_raw.filter(f.col('comments').isNotNull())
```

c) Cleaning of the prices in listings

```python
#how many listings do we have ?
listings_raw.count()
```

```python
#how many listing prices start with $? are there multiple currencies?
listings_raw.filter(f.col('price').startswith('$')).count()
```

```python
#let's check those that don't start with $
listings_raw.filter(~(f.col('price').startswith('$'))).select('price').show()
```

```python
#we don't know the currency of the ones that don't have the $ sign. 2 don't make sense for a price, and we wouldn't know how to
#convert the other one. So instead of introducing potentially false values, let's drop it
listings_raw =listings_raw.filter(f.col('price').startswith('$'))
```

```python
#Now let's turn the prices into numerical values so that we can use them in queries
listings_raw = listings_raw.withColumn("price", f.translate(f.col("price"), "$,", "").cast('float'))
```

d) Cleaning of the date in reviews

```python
#the date attribute is crucial to the information, and from the data that we have there is no recover it,
#therefore, we are dropping the records that have null as date
reviews_raw = reviews_raw.filter(f.col('date').isNotNull())
```

```python
#let's check the format of the dates to see if we can cast them into dates without introducing nulls
#it will introduce 95 nulls, however a review is only relevant if it's dated. And there is no way to recover this information.
#also, the number of nulls is insignicant (95 out of more than a million reviews)
reviews_raw.filter(f.year(f.col('date')).isNull()).count()
```

```python
#the colab storage capacity is exceeded if we run the reviews with the whole dataset
#therefore in this notebook we run the following part using the reduced datatset (100k lines)
#to run it on the whole dataset, comment the two following lines
listings_raw = listings_raw.limit(100000)
reviews_raw = reviews_raw.limit(100000)
```

```python
### test tokenizer
# augment the reviews dataframe with the ratings contained in the listings dataframe, joining on the base of the id of the listings.
# in this way, each review is associated to the corresponding listing through the listing id.
# at the end, retain the comment, the id of the listing and the rating.
reviews_aug = reviews_raw.select('listing_id','comments').join(listings_raw.select('id','review_scores_rating'),reviews_raw.listing_id == listings_raw.id).select(listings_raw.review_scores_rating,\
                                                                                                                                                  reviews_raw.comments,\
                                                                                                                                                  reviews_raw.listing_id)
reviews_aug.count()
# use tokenize to divide the words in the comments.                                                       
tokenizer = Tokenizer(inputCol="comments", outputCol="words")

tokenized = tokenizer.transform(reviews_aug.select('comments'))

# use StopWordsRemover to remove the words that do not carry meaning, such as "the", "is"...
remover = StopWordsRemover(inputCol="words", outputCol="filtered")
removed = remover.transform(tokenized)
# now, the dataframe "removed" contains three columns: the full comment, the comment with words divided and the comment with words divided
# with the stop words removed. we join the dataframe "removed" to the original reviews_aug dataframe on the base of the comments.

words_split = reviews_aug.join(removed, reviews_aug.comments == removed.comments)\
              .select(reviews_aug.review_scores_rating, removed.filtered,reviews_aug.listing_id).distinct()

# now, we need to explode the dataframe in order to have a dataframe containing all the single words along with the rating and the listing in
# which they compare. note that, for now, the words are repeated in the words_split dataframe: if a word appears in multiple listings, it
# appears multiple times in the words_split dataframe. later, the words will be grouped.
words_split = words_split.select('review_scores_rating', f.explode('filtered').alias('word'),'listing_id')

# some words still contain characters like punctuation. ise regexp_replace to remove these characters.
words_split = words_split.withColumn('word', f.regexp_replace(f.col('word'),"[,|.|!|'|\|/|(|)|?|:|;|-|_]",""))

# now, we want the words to compare only one time in the dataframe, and to group them on the base of some useful parameters.
# the parameters that are considered are three: the mean of the score associated to each word (i.e. the mean of the ratings of the 
# listings in which the word appear), the number of times the word appears (i.e. frequency), and the number of unique reviews in which
# the word appears.
words_split_grouped = words_split.groupBy('word').agg(f.avg('review_scores_rating').alias('average_rating'),f.count("*")\
                                                      .alias('frequency'),f.countDistinct('listing_id').alias('unique_listings'))\
                                  .orderBy('average_rating', ascending = True)

# now, in some way these three parameters should be used to calculate a score function. since these parameters have different orders of
# magnitude, they should be normalized, so they can be compared.
# to normalize, we use MaxAbsScaler, that takes a column as input and returns the normalized column.
# however, we need to normalize three columns (ratings, frequency and number of unique listigns), and not only one. hence, we use VectorAssembler
# that combines the three column in a single column (named "features"), which can be given as input to MaxAbsScaler.
# MaxAbsScaler returns a column "scaledFeatures" that contains, for each row, a vector of three elements: the first element is the 
# normalized rating, the second element is the normalized frequency, the third element is the normalized number of unique listings in which
# the word occurs.
assembler = VectorAssembler(
    inputCols=["average_rating", "frequency", "unique_listings"],
    outputCol="features")
output = assembler.transform(words_split_grouped)
#output.show()
scaler = MaxAbsScaler(inputCol="features", outputCol="scaledFeatures")
scalerModel = scaler.fit(output)
scaledData = scalerModel.transform(output)
#scaledData.show()

def score_fcn(feature,unique_listings):
  score = float(3*feature[0]) + float(feature[1]) + float(feature[2])
  if unique_listings < 100:
    score = None
  return(score)

score_fcn_udf = f.udf(score_fcn)

# define a list of words that we think should be ignored (i.e. that appear for some reason in best/worst word but are not strictly related to that)
ignored_words = ['apartment','stockholm','place','host','location','city','us', '']
    
# apply the function and drop the words that have null score. filter also the words in ignored_words
words_rated = scaledData.where(~scaledData.word.isin(ignored_words)).withColumn('score',score_fcn_udf(f.col('scaledFeatures'),f.col('unique_listings'))\
                      .cast(FloatType())).where(f.col('score').isNotNull()).select('word','score')

# best 10 words
print('The best 10 words are:')
words_rated.orderBy('score', ascending = False).show(10)
# worst 10 words
print('The worst 10 words are:')
words_rated.orderBy('score', ascending = True).show(10)
```

**Comment :**

We can see that the top 10 words and the bottom 10 words make sense in terms of rating as they are words that we would naturally associate with good or bad connotations.

```python
#Let's look at the most popular part per country each year
number_sold_per_year = spark.sql('select part, country, year(date) as year, count(*) as number_sold from transactions group by year, part, country')

windowSpec  = Window.partitionBy('country', 'year').orderBy(f.col('number_sold').desc())

number_sold_per_year.createOrReplaceTempView("number_sold_per_year")

popular_parts_per_country = number_sold_per_year.withColumn("rank",rank().over(windowSpec)).filter(f.col('rank')<2).select('country', 'part', 'year', 'number_sold')

most_popular_parts_country = popular_parts_per_country.orderBy('country','year').groupBy('country','year').agg(f.first('part').alias('part')).groupBy("country")\
.pivot("year")\
.agg(f.first("part"))

#showing the table
#We can first show a prettier array for the top five companies: 
fig = plt.figure(figsize = (11,6))
ax = fig.add_subplot(111)

table = ax.table(cellText = most_popular_parts_country.toPandas().values,
          colLabels = ['Country', '2016', '2017', '2018'],
          loc = "center",
          colColours =["skyblue"] * 10,
          )
table.set_fontsize(20)
table.scale(1, 2)
ax.set_title("Most popular parts in each country throughout the years ", fontweight ="bold")

ax.axis("off")
```

**Comments :**

We can see that the favorite product in each country each year changes. There is no consecutive years where the same part has been the most popular in a country.

There is also a diversity in terms of part highest in demand between the countries.

Therefore we can see that our customers have different needs depending on the countries they are in, and their needs also vary with years.
